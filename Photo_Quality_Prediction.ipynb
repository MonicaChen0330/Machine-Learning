# 2 Logistic Regression

## 2.1 Default Logistic Regression

### 2.1.1 Build Model and Make Prediction

```python
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression() # with default hyper-parameters

logReg.fit(x_tr, y_tr)

y_pr = logReg.predict(x_te)
```

### 2.1.2 Evaluate Result

```python
# view confusion matrix

from sklearn.metrics import confusion_matrix

cfs = confusion_matrix(y_te, y_pr)

fig, ax = plt.subplots(figsize=(2.5, 2.5))
ax.matshow(cfs, cmap=plt.cm.Blues, alpha=0.3)
for i in range(cfs.shape[0]):
 for j in range(cfs.shape[1]):
  ax.text(x=j, y=i,
  s=cfs[i, j],
  va='center', ha='center')
plt.xlabel('predicted label')
plt.ylabel('true label')
plt.show()

print("True Positive: ", cfs[0, 0])
print("True Negative: ", cfs[1, 1])
print("False Positive: ", cfs[0, 1])
print("False Negative: ", cfs[1, 0])
```

```python
# evaluate with scoring

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

print("Accuracy_Tr: ", logReg.score(x_tr, y_tr).round(5))
print("Accuracy_Te: ", accuracy_score(y_te, y_pr).round(5))
print("Precision: ", precision_score(y_te, y_pr).round(5))
print("Recall: ", recall_score(y_te, y_pr).round(5))
print("F1 Score: ", f1_score(y_te, y_pr).round(5))
```

## 2.2 Grid-Search for Hyper Parameters

### 2.2.1 Execute Grid Search

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV


model = Pipeline([
    ('logReg', LogisticRegression())
])

param = [{
    'logReg__penalty': ['l1', 'l2'],
    'logReg__solver': ['lbfgs', 'liblinear']
}]

score = 'accuracy'
```

```python
gridSearch = GridSearchCV(
    estimator = model,
    param_grid = param,
    scoring = score,
    cv = 10)

gridSearch.fit(x_tr, y_tr)
```
